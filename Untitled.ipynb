```python
label_paths = glob.glob(dataset_path + "labels/*")
label_paths.sort()
label_paths[:5]
```

```python
from scipy import io
paths=[]
names=[]
for i in label_paths:
    matdata = io.loadmat(i, squeeze_me=True)
    
    if not matdata["action"] in names:
        names.append(matdata["action"])
    if matdata["action"] == "tennis_serve":
        paths.append(i)
    
```
